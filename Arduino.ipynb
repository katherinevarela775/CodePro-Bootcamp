```python
##Codigo semaforo con Fade
int rojo = 9;
int amarillo = 10;
int verde = 11;

void setup() {
  pinMode(rojo, OUTPUT);
  pinMode(amarillo, OUTPUT);
  pinMode(verde, OUTPUT);
}

// Función auxiliar para desvanecer una luz
void efectoFade(int pin) {
  // Aumentar brillo gradualmente
  for (int i = 0; i <= 255; i += 5) {
    analogWrite(pin, i);
    delay(30); // Velocidad del desvanecimiento
  }
  delay(1000); // Se mantiene encendido un segundo
  // Disminuir brillo gradualmente
  for (int i = 255; i >= 0; i -= 5) {
    analogWrite(pin, i);
    delay(30);
  }
}

void loop() {
  efectoFade(rojo);
  efectoFade(amarillo);
  efectoFade(verde);
}
```

```python
##Codigo semaforo con villancicos
// Definición de notas para Jingle Bells
#define NOTE_E4  330
#define NOTE_G4  392
#define NOTE_C4  262
#define NOTE_D4  294

int buzzer = 8;
int rojo = 9, amarillo = 10, verde = 11;

// Melodía y duración de notas
int melody[] = { NOTE_E4, NOTE_E4, NOTE_E4, NOTE_E4, NOTE_E4, NOTE_E4, NOTE_E4, NOTE_G4, NOTE_C4, NOTE_D4, NOTE_E4 };
int durations[] = { 4, 4, 2, 4, 4, 2, 4, 4, 4, 4, 1 };

void setup() {
  pinMode(rojo, OUTPUT);
  pinMode(amarillo, OUTPUT);
  pinMode(verde, OUTPUT);
}

void loop() {
  // --- FASE VERDE ---
  digitalWrite(verde, HIGH);
  tocarMelodia();
  digitalWrite(verde, LOW);

  // --- FASE AMARILLA ---
  digitalWrite(amarillo, HIGH);
  delay(1000); 
  digitalWrite(amarillo, LOW);

  // --- FASE ROJA ---
  digitalWrite(rojo, HIGH);
  tocarMelodia();
  digitalWrite(rojo, LOW);
}

void tocarMelodia() {
  for (int i = 0; i < 11; i++) {
    int duration = 1000 / durations[i];
    tone(buzzer, melody[i], duration);
    
    // El brillo del LED actual oscila levemente con el ritmo
    int pausa = duration * 1.30;
    delay(pausa);
    noTone(buzzer);
  }
}
```
