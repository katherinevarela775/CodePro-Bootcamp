Si la nota es 90 o más, imprimir "Excelente, exonerado".

Si está entre 70 y 89, imprimir "Aprobado".

Si está entre 60 y 69, imprimir "Derecho a recuperatorio".

Si es menor a 60, imprimir "Reprobado".

Extra (Lo solicitado del final del día 2): Envuelve todo lo anterior en un bucle while para que el programa siga pidiendo notas hasta que el usuario escriba el número 0 para salir.

```python
#Sistema de Notas
nota = 70

while nota != 0 :
    nota = int(input(f'Ingrese la nota del estudiante'))
    print (f'La nota ingresada fue {nota}')
    if nota >= 90:
        print('Excelente, exonerado')
    elif 70<= nota <= 89:
        print('Aprobado')
    elif 60< nota <=69:
        print ('Derecho a recuperatorio')
    elif nota < 60:
        print('Pa febrero')
```

El Verificador de Edad para Eventos: Escribe un programa que use un bucle infinito (while True) para pedir edades.

Si la edad es menor a 13, imprimir "Entrada prohibida".

Si la edad está entre 13 y 17, imprimir "Entrada con tutor".

Si la edad es 18 o más, imprimir "Entrada libre".

Condición de salida: Si el usuario ingresa la palabra "salir", el programa debe terminar usando break. Pista: Tendrás que validar si el input es "salir" antes de convertirlo a número.


```python
#Verificador de edad para eventos

edad = 30

while True:
    edad = input('Ingrese su edad')
    if edad == 'salir':
        print('El programa ha finalizado')
        break
    elif int(edad) < 13:
        print('Entrada prohibida')
    elif 13 <= int(edad) <= 17:
        print('Entrada con tutor')
    elif int(edad) >= 18:
        print('Entrada libre')
    
```
