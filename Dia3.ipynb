## Challenge 3 游뱤 - 15m

Crear una funci칩n que reciba `dos par치metros` (una palabra y un n칰mero), y que retorne `True` si el `n칰mero es igual` a la cantidad de letras de la palabra. En caso contrario, que retorne `False`.

Pista: `cantidad de letras de una palabra en python`

````python

def caja(palabra, num):
    if len(palabra) == num:
        return True
    else:
        return False

palabra_ingresada= input("Ingrese una palabra")
numero_ingresado= int(input('Ingrese un numero'))

caja(palabra_ingresada, numero_ingresado)
````

## Challenge 4 游뱤 - 20 min
Crear `funci칩n` que revise si una palabra `contiene` una letra. En caso de que la contenga, `retornar` la letra entre corchetes `[]`. De lo contrario, `retornar` solo la letra.

````python
letraa= input('Ingrese una letra')
palabraa= input('Ingrese una palabra')

def buscador(letra, palabra):
    print(f'La palabra ingresada fue: {palabra}')
    for i in palabra:
        if i == letra:
            return [letra]
    return("La palabra no contiene la letra")


buscador(letraa, palabraa)
````
