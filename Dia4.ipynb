## Challenge 2 ðŸ¤º - 20 min

Crear una tarjeta personal. El usuario (mediante `input`) puede elegir:

1. `Crear` datos nuevos
2. `Modificar` datos
3. `Eliminar` datos

```python
tarjeta_personal = {}
print(tarjeta_personal)
print('Para finalizar ingrese la palabra "salir"')

accion_deseada = "existir"


while True:
    accion_deseada = input('Ingrese la accion que desea realizar')

    if accion_deseada == "salir":
        print('Se han ingresado todos lo datos')
        break
    elif accion_deseada == 'crear':
        dato_usuario = input('Ingrese el dato que desea crear')
        valor_dato = input('Asignele un valor a ese dato')
        tarjeta_personal[dato_usuario]= valor_dato
        print(tarjeta_personal)

    elif accion_deseada == 'modificar':
        dato_usuario = input('Ingrese el dato que quiere modificar')
        valor_dato = input('Dele un nuevo valor a ese dato')
        tarjeta_personal[dato_usuario]= valor_dato
        print(tarjeta_personal)

    elif accion_deseada == 'eliminar':
        eliminar_item = input('Ingrese el dato que desea eliminar')
        del tarjeta_personal[eliminar_item]
        print(tarjeta_personal)

    
```

## Challenge 3 ðŸ¤º
Crear una agenda de contactos. El usuario puede elegir:

1. `Mostrar` los contactos
2. `Crear `contacto
3. `Modificar` datos del contacto

```python
agenda_de_contactos = {}
print(agenda_de_contactos)
print('Para finalizar ingrese la palabra "salir"')

accion_a_realizar = "existir"


while True:
    accion_a_realizar = input('Ingrese la accion que desea realizar')

    if accion_a_realizar == "salir":
        print('Se han ingresado todos los contactos')
        break
    elif accion_a_realizar == 'crear':
        dato_contacto = input('Ingrese el nombre del contacto')
        valor_contacto = input('Ingrese el numero de contacto')
        agenda_de_contactos[dato_contacto]= valor_contacto
        print(agenda_de_contactos)

    elif accion_a_realizar == 'mostrar':
        dato_contacto = input('Ingrese el dato que quiere mostrar')
        print(agenda_de_contactos[dato_contacto])

    elif accion_a_realizar == 'modificar':
        dato_contacto = input('Ingrese el dato que quiere modificar')
        valor_contacto = input('Dele un nuevo valor a este contacto')
        agenda_de_contactos[dato_contacto]= valor_contacto
        print(agenda_de_contactos)
   
```
