```python

def articulo_a_comprar(articulo, tipo, objetivo):
    print("Buenas Tardes, que articulo necesitaria")
    print(f'Necesito una {articulo} de {tipo} para arreglar mi {objetivo}')
    print('-----------------------------------------------')

especif1= input('Defina el articulo de su eleccion')
especif2= input('Especifique la medida o tipo')
especif3= input('Çuentenos para que lo necesita')

articulo_a_comprar(especif1, especif2, especif3)

articulo_a_comprar('LLave', '3/4', 'moto')
articulo_a_comprar('Lija', '100', 'puerta')
articulo_a_comprar('pintura', 'exterior', 'fachada')
```
